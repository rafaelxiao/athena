```python
s = ts.get_report_data(2016, 4)
```

```python
df = s[['code', 'name', 'distrib']]
```

```python
def decompose_distrib_list(text_string, signal):
    accept_range = [str(i) for i in range(10)] + ['.']
    idx = -1
    try:
        idx = text_string.find(signal)
    except:
        pass
    if idx >= 0:
        value = ''
        for i in text_string[idx+1:]:
            if i in accept_range:
                value += i
            else:
                break
        value = float(value)
    else:
        value = 0.0
    return value
```

```python
df['dividend'] = df.apply(lambda row: decompose_distrib_list(row['distrib'], '派'), axis=1)
df['bonus'] = df.apply(lambda row: decompose_distrib_list(row['distrib'], '送'), axis=1)
```

```python
df = df.dropna()
```
